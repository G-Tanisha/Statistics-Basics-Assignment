1.  What is the difference between descriptive statistics and inferential statistics? Explain with examples.
    - Descriptive statistics and inferential statistics are two key branches of statistics that serve different purposes. Descriptive statistics involve methods for summarizing and organizing data so that it can be easily understood. This includes measures like mean, median, mode, range, standard deviation, and the use of charts or graphs. For example, if a teacher calculates the average score of a class test and shows the score distribution using a histogram, they are using descriptive statistics. On the other hand, inferential statistics go beyond merely describing the data and are used to make predictions or generalizations about a larger population based on a sample. This involves techniques like hypothesis testing, confidence intervals, and regression analysis. For instance, if a survey is conducted on 200 people to predict the political preference of an entire city, the conclusions drawn using that sample fall under inferential statistics. In short, descriptive statistics describe what the data shows, while inferential statistics help in making informed guesses or decisions about a broader group based on limited data.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
  - Sampling in statistics is the process of selecting a subset of individuals or items (called a sample) from a larger group or population in order to analyze and draw conclusions about the entire population without having to study every member of it. Sampling is essential because studying an entire population is often impractical, expensive, or time-consuming.
  There are different methods of sampling, and two common types are random sampling and stratified sampling.

  In random sampling, every individual in the population has an equal chance of being selected. This method is simple and reduces selection bias. For example, if you randomly pick 50 students from a college of 500 using a lottery system, it is random sampling.
  
  In contrast, stratified sampling involves dividing the population into distinct subgroups or strata based on a certain characteristic (like age, gender, income level), and then randomly selecting samples from each subgroup proportionally. This method ensures that each subgroup is fairly represented in the final sample. For instance, if a college has 60% science students and 40% arts students, and you want a sample of 100 students, stratified sampling would involve selecting 60 science students and 40 arts students randomly from their respective groups.

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.
   - Mean, median, and mode are the three main measures of central tendency in statistics, used to describe the center or average of a dataset. They help summarize and understand the general behavior of data.
   Mean is the arithmetic average of a set of numbers. It is calculated by adding all the values and dividing by the number of values. For example, the mean of 2, 4, and 6 is (2 + 4 + 6) ÷ 3 = 4. It gives a general idea of the overall level of the data, but it can be affected by extreme values (outliers).
   Median is the middle value when the data is arranged in order. If there is an even number of values, it is the average of the two middle numbers. For example, in the set 3, 5, 7, the median is 5. In the set 2, 4, 6, 8, the median is (4 + 6) ÷ 2 = 5. The median is useful because it is not affected by outliers and gives a better sense of the central location in skewed data.Mode is the value that appears most frequently in a dataset. For example, in the set 2, 3, 3, 4, 5, the mode is 3. A dataset can have one mode, more than one mode, or no mode at all. Mode is helpful for identifying the most common or popular value, especially in categorical data.
   These measures of central tendency are important because they provide a quick summary of a dataset and help in making comparisons, understanding distributions, and guiding decision-making based on data trends.

4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
  - Skewness and kurtosis are statistical measures used to describe the shape of a data distribution. Skewness refers to the asymmetry of the distribution. If the data is symmetrically distributed, skewness is zero. A positive skew (right-skewed) means the tail on the right side is longer, indicating that most values are concentrated on the lower end, with a few extremely high values pulling the mean above the median. Conversely, a negative skew has a longer left tail with most values on the higher side. Kurtosis, on the other hand, measures the "tailedness" or the sharpness of the peak of the distribution. A normal distribution has moderate kurtosis, while a distribution with high kurtosis (leptokurtic) has a sharp peak and heavy tails, suggesting more outliers. A distribution with low kurtosis (platykurtic) has a flatter peak and lighter tails, indicating fewer outliers. Together, skewness and kurtosis help in understanding the shape and behavior of data beyond basic measures like the mean and standard deviation.





```python
'''Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]'''
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
'''Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]'''
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y, bias=False)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
'''You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]'''

'''Covariance tells us the direction of the relationship between two variables.A positive covariance means that when advertising spend increases, daily sales also tend to increase.A negative covariance suggests that as advertising spend increases, daily sales tend to decrease.
However, covariance does not tell us the strength of the relationship or standardize it.
Correlation (specifically the Pearson correlation coefficient) not only tells us the direction but also the strength of the relationship.
It ranges from -1 to 1:
1: Perfect positive correlation
0: No correlation
-1: Perfect negative correlation'''


import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

covariance = np.cov(advertising_spend, daily_sales, bias=False)[0][1]
correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
'''Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]'''
'''To understand the distribution of customer satisfaction survey scores (on a scale of 1–10),use the following summary statistics and visualizations:
Mean: Average score — gives a general sense of satisfaction.
Median: Middle value — shows central tendency without being affected by outliers.
Mode: Most frequent score — helps identify the most common rating.
Standard Deviation: Measures how spread out the scores are — lower values indicate consistency.
Minimum and Maximum: Helps identify range and possible outliers.
Visualizations:
Histogram: Shows how frequently each score appears — useful for understanding the shape of the distribution (e.g., skewed, normal, bimodal).
Boxplot: Can be used to detect outliers and understand the data spread.'''
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=7, edgecolor='black', color='skyblue')
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Score (1-10)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
